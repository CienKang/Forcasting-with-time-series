```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("/content/DATASET.xlsx")
data
```

```python
# since 1009 is not divisible by 10
# Therefore one section must have less than 10 values
# Which on manual checking was found to be 9
```

```python
# Removing Unamed:0 columns 
data = data.reset_index()
data.drop(columns=['Unnamed: 0','index'],inplace=True)
data
```

```python
# Creating different csv files for different sections
sec = 1
i = 0
while i < len(data):
  end = 10
  if i == 910:
    end = 9

  new_df = pd.DataFrame(columns=['year',	'Para-1',	'Para-2',	'Para-3',	'Para-4',	'Para-5',	'Para-6',	'Para-7',	'Para-8',	'Para-9',	'Para-10',	'Para-11',	'Para-12',	'Para-13'])
  for j in range(0, end):
    row = data.loc[i+j, :]
    new_df = new_df.append(row)
  
  if i == 910:
    i = i+9
  else :
    i = i +10
  new_df.to_csv("Sections/section"+str(sec)+".csv", index=False)
  sec = sec +1
```

```python
from math import sqrt
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
RMSE = 0
```

```python

def make_predictions(df,section):
  cols = ['Para-9', 'Para-10',	'Para-11',	'Para-12',	'Para-13']
  x = df.drop(columns=cols, axis=1)
  y = df[cols]
  
  x_train, x_test = x.iloc[:-1, :], x.iloc[-1:, :]
  y_train, y_test = y.iloc[:-1], y.iloc[-1:]

  model = LinearRegression()
  model.fit(x_train, y_train)
  predictions = model.predict(x_test)
  rmse = np.round(sqrt(mean_squared_error(y_test, predictions)), 3)
  return rmse
```

```python
i = 1
while i <= sec:
  df =pd.read_csv('Sections/section'+str(i)+'.csv')
  RMSE = RMSE + make_predictions(df,i)
  i = i +1
```
